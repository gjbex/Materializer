## Mask

A mask can be defined, 255 (`#FFFFFF`) is considered masked out.

```python
width, height = 1000, 1000
```

```python
x, y = np.ogrid[:width, :height]
mask = (x - width//2)**2 + (y - height//2)**2 > (0.45*width)**2
mask = 255*mask.astype(int)
```

```python
plt.imshow(mask)
```

```python
mask[0, 0], mask[width//2, height//2]
```

## Word cloud

The layout of the word cloud is random, so you may want to run it multiple times until you get something pleasing.

```python
cloud = wordcloud.WordCloud(width=1000, height=1000, max_words=100, repeat=False, background_color='white', mask=mask).generate_from_frequencies(keywords)

figure, axes = plt.subplots(figsize=(15, 15))
axes.imshow(cloud, aspect='equal')
axes.get_xaxis().set_visible(False)
axes.get_yaxis().set_visible(False)
```

Don't forget to save the result to a file.

```python
figure.savefig('test.jpeg')
```

# C++

## Words

```python
c_plus_plus_keywords = ['int', 'uint64', 'long', 'float', 'double', 'in68', 'size_t', 'char', 'bool', 'const', 'void', 'unsigned', 'size_t',
                        'struct', 'class', 'template', 'namespace', 'using',
                        '#include', '#ifdef', '#endif', '#ifndef', '#else', '#define',
                        'std::string', 'std::list', 'std::map', 'std::set', 'std::unordered_set', 'std::vector', 'std::array', 'std::valarray',
                        'std::begin', 'std::end', 'std::cbegin', 'std::cend', 'std::endl',
                        'std::unique_ptr', 'std::shared_ptr', 'new', 'delete',
                        'if', 'else', 'for', 'while', 'break', 'continue', 'switch', 'case', 'try', 'catch', 'return',
                        'std::iota', 'std::sort', 'std::find', 'std::foreach', 'std::shuffle', 'std::copy', 'std::move', 'std::rotate', 'std::generate_n', 'std::replace_if',
                        'std::fill', 'std::reverse', 'std::unique', 'std::min', 'std::max', ]
```

```python
meta_keywords = ['STL', 'HPC', 'exception', 'object', 'algorithm', ]
```

```python
keywords = {keyword: random.randint(8, 16) for keyword in c_plus_plus_keywords}
keywords.update({keyword: random.randint(12, 20) for keyword in meta_keywords})
```

## Mask

```python
c_plus_plus_img = np.genfromtxt('AsciiImages/c_plus_plus.txt')
```

```python
c_plus_plus_mask = 255*(1 - c_plus_plus_img).astype(np.int32)
```

## Word cloud

```python
cloud = wordcloud.WordCloud(width=1000, height=1000, max_words=100, repeat=False, colormap='copper', background_color=None, mode='RGBA', mask=c_plus_plus_mask).generate_from_frequencies(keywords)

figure, axes = plt.subplots(figsize=(15, 15))
axes.imshow(cloud, aspect='equal')
axes.get_xaxis().set_visible(False)
axes.get_yaxis().set_visible(False)
```

```python
figure.savefig('c_plus_plus.jpeg')
```
